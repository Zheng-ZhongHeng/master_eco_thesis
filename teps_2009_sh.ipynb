```python

ax = df["contact"].value_counts().plot(kind='bar', figsize=(12, 9), rot=20, fontsize=12,) 
for i in ax.containers:
    ax.bar_label(i, label_type='edge')
    plt.xticks(rotation=20)
plt.show()
count_contact_sh_2009=pd.DataFrame(df["contact"].value_counts())
count_contact_sh_2009
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot') 
for col in df.columns:
    if df[col].dtype == 'object': # check if the column contains categorical data
        ax = df[col].value_counts().plot(kind='barh', figsize=(12, 9), rot=20, fontsize=12, colormap='cividis') 
        for i in ax.containers:
            ax.bar_label(i, label_type='edge')
        plt.xticks(rotation=20)
        plt.title(col, fontsize=16) # set the title of the plot to the column name
        plt.tight_layout()
        plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=len(df.columns), figsize=(300, 200))

for i, col in enumerate(df.columns):
    if df[col].dtype == 'object': # check if the column contains categorical data
        df[col].value_counts().plot(kind='barh', ax=ax[i], fontsize=12, colormap='cividis') 
        ax[i].set_title(col, fontsize=16) # set the title of the plot to the column name
        ax[i].tick_params(axis='y', which='major', pad=10) # increase the distance between the tick labels and the axis
        for j in ax[i].containers:
            ax[i].bar_label(j, label_type='edge')
plt.tight_layout()
plt.show()
```
