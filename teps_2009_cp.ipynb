```python
# Data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 設定中文字體
plt.rcParams['font.family'] = ['Arial Unicode MS']
font_path = '/System/Library/Fonts/PingFang.ttc'
font_prop = fm.FontProperties(fname=font_path, size=16)
plt.style.use('ggplot')

data = {
    'TEPS-B 樣本性別': ['男', '女', 'Total'],
    '教育領域': [38, 50, 88],
    '人文及藝術領域': [135, 322, 457],
    '社會科學、商業及法律領域': [261, 494, 755],
    '科學領域': [234, 136, 370],
    '工程、製造及營造領域': [574, 125, 699],
    '農學領域': [31, 43, 74],
    '醫藥衛生及社福領域': [78, 171, 249],
    '服務領域': [118, 158, 276],
    '其他領域': [3, 3, 6],
    '遺漏值': [173, 224, 397],
    '不知道、不清楚': [9, 8, 17],
    '拒答': [6, 2, 8],
    '跳答': [104, 93, 197],
    'Total': [1764, 1829, 3593]
}

df = pd.DataFrame(data)

# Plotting pie charts for both genders with subplots

fig, axes = plt.subplots(1, 2, figsize=(12, 9))

# Data
labels = df.columns[1:-1]
sizes_men = df.iloc[0, 1:-1]
sizes_women = df.iloc[1, 1:-1]

# Men's data
axes[0].pie(sizes_men, labels=labels, autopct='%1.1f%%', startangle=140)
axes[0].set_title('男生', fontproperties=font_prop)

# Women's data
axes[1].pie(sizes_women, labels=labels, autopct='%1.1f%%', startangle=140)
axes[1].set_title('女生', fontproperties=font_prop)

plt.tight_layout()
plt.show()
```
