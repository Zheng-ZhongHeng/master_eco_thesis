```python
import pandas as pd

# Create an empty DataFrame to store the results
results_df = pd.DataFrame()

# Loop through each column in the DataFrame
for col in df.columns:
    # Check if the column contains numerical data
    #if df[col].dtype != 'object':
        # Calculate the descriptive statistics for the column
        stats = df[col].describe()
        # Add the column name as a new row index
        stats.name = col
        # Transpose the result and add it to the results DataFrame
        results_df = pd.concat([results_df, stats.to_frame().T])

# Print the results DataFrame
print(results_df)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot') 
for col in df.columns:
    if df[col].dtype == 'object': # check if the column contains categorical data
        ax = df[col].value_counts().plot(kind='barh', figsize=(12, 9), rot=20, fontsize=12, colormap='plasma') 
        for i in ax.containers:
            ax.bar_label(i, label_type='edge')
        plt.xticks(rotation=20)
        plt.title(col, fontsize=16) # set the title of the plot to the column name
        plt.tight_layout()
        plt.savefig(col+'.png',dpi=10000, )
        plt.show()
```

```python
# Define a dictionary of code-label mappings
code_labels_contact = {0: '未成功接觸', 1: '成功接觸未完訪', 2: '成功完訪', 9: '無法再接觸'}

# Replace the codes with labels in the 'contact' column
cp2013['contact'] = cp2013['contact'].replace(code_labels_contact)
cp2013['contact'].value_counts()    
```

```python
cp2013['tepsb_sex'].value_counts()
code_labels_sex={1:'男',2:'女'} 
cp2013["tepsb_sex"]=cp2013["tepsb_sex"].replace(code_labels_sex)
cp2013["tepsb_sex"].value_counts()
```
